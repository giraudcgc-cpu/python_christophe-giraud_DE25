5. Monte Carlo simulation



```python

# Copied an paste from grok

import numpy as np
import matplotlib.pyplot as plt

# Set random seed for reproducibility
np.random.seed(42)

# Generate 5000 points with x, y ~ U(-1, 1)
n_points = 5000
x = np.random.uniform(-1, 1, n_points)
y = np.random.uniform(-1, 1, n_points)

# Calculate Euclidean distances from origin
distances = np.sqrt(x**2 + y**2)

# Create masks for points inside and outside the unit circle
inside = distances < 1
outside = distances > 1

# Create scatter plot
plt.figure(figsize=(8, 8))
plt.scatter(x[inside], y[inside], c='blue', s=10, label='d < 1', alpha=0.6)
plt.scatter(x[outside], y[outside], c='red', s=10, label='d > 1', alpha=0.6)

# Add unit circle for reference
circle = plt.Circle((0, 0), 1, color='black', fill=False, linestyle='--')
plt.gca().add_artist(circle)

# Set plot properties
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Scatter Plot of 5000 Points with Euclidean Distance Coloring')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.gca().set_aspect('equal')  # Ensure equal scaling for circle

# Show plot
plt.show()
```

h) What is the difference between sorted() and .sort() in lists? <br>
sorted(): a function that takes any iterable (e.g., list, tuple, string) as input and returns a new sorted list. The original iterable remains unchanged.<br>
.sort(): A method that sorts the list in place, modifying the original list. It does not return a new list.

ex: <br>
my_list = [3, 1, 4, 2] <br>
new_list = sorted(my_list) <br>
print(new_list)      # Output: [1, 2, 3, 4] <br>
print(my_list)       # Output: [3, 1, 4, 2] (original unchanged)


ex: <br>
my_list = [3, 1, 4, 2] <br>
my_list.sort() <br>
print(my_list)       # Output: [1, 2, 3, 4] (original list modified) <br>
print(my_list.sort()) # Output: None (returns None, not a new list) <br>
