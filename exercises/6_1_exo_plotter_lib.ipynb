PLOTTER LIBRARY

```python
import numpy as np
import matplotlib.pyplot as plt

class AnnotatePlot:
    pass

np.random.seed(42)
x = np.linspace(0,10,100)
y = 2 * x + 5 + np.random.normal(0,2, size=len(x))

print(f"{x[:5] = }")
print(f"{x.shape =}")

# better way to do it:
fig, ax = plt.subplots(1)

plt.plot(x, y)
ax.set(title = "Cool stock", xlabel = "day", ylabel = "SEK")
ax.spines[["top", "right"]].set_visible(False)
ax.annotate("Big dip here", xytext = (9, 10), xy =(7.5, 14.2), arrowprops={"arrowstyle":"->"})
```

```python
annotate = AnnotatePlot(fig, ax)
annotate.arrow("Big dip here", (8, 10), (7.5, 14.8))
```

REFACTOR INTO A CLASS

```python
class PlotAnnotator:
    def __init__(self, fig, ax):
        self.fig = fig
        self.ax = ax

    def arrow(self, label:str, start_xy:tuple[float], end_xy: tuple[float]):
        self.ax.annotate(
            label, 
            xytext = (9, 10),
            xy =(7.5, 14.2), 
            arrowprops={"arrowstyle":"->"}
            )
        return self

    def remove_clutter(self):
        self.ax.spines[["top", "right"]].set_visible(False)
        return self

fig, ax = plt.subplots(1)

ax.plot(x, y)

annotator = PlotAnnotator(fig, ax)

annotator.arrow("Big dip here", start_xy= [9, 10], end_xy = (7.5, 14.2))
```
