- focus on the distinction between 2 ways of writting matplotlib codes
- focus on the components of a matplotlib figure

```python
import matplotlib.pyplot as plt  # Here I want to use the pyplot module in the matplotlib package
import numpy as np
```

```python
# Generate data
x = np.linspace(-5, 5, 100)
x
# I can print it to see the array of numbers
```

```python
# Generate data f
f = np.sin(x)
f
# Same here, if I print it I see the array of nbers for f
```

#### 1rst approach - functional/matlab style

```python
plt.figure()
plt.plot(x,f)
plt.title("sin(x)")
plt.xlabel("X")
plt.ylabel("F")
```

#### 2nd approach - OOP

```python
fig = plt.figure()      # this createa figure
ax = plt.axes()         # this creates an axe
ax.plot(x, f)           # create a plot
ax.set(xlim=(-10, 10), ylim= (-5, 5), xlabel="X", ylabel="Y")       # to modify the axe object

# check more at https://matplotlib.org/stable/gallery/showcase/anatomy.html (just curiosity)
```
